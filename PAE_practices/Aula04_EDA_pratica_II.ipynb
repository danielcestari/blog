```python
from matplotlib import pyplot as plt
iris.columns[[0,1,2,3]]
```

```python
# comprimento da pétala maior e menor que 1.12, 1.8 na classe 0.0
iris[(iris.target == 0.0) & ((iris.iloc[:, 2] < 1.2) | (iris.iloc[:, 2] >1.8))]
# também não parece ser um outlier
```

**Respondendo perguntas sobre momentos**

- Quando é indicado utilizar os momentos estatísticos para EDA?
    
    Os momentos são descritores da distribuição dos dados, na verdade se você especificar todos os momentos você acaba especificando qual distribuição exatamente você está trabalhando. Logo, os momentos são indicados quando você precisa de uma maneira descrever a distribuição. Tipicamente, média e variância (ou desvio padrão) são usadas, mas isso assume que a distribuição segue uma normal e essas duas medidas descrevem bem essa distribuição. Mas há outras distribuições que não são bem descritas pela média e distribuição.
