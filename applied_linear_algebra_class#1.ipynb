### Transformações lineares

$T:\mathbb{R}^n \to \mathbb{R}^m$


$T(\alpha x + y ) = \alpha T(x) + T(y)$

$T(0)=0$

Existe uma relação biunívoca entre uma transformação linear e uma representação matricial para essa transformação.

$A_{m \times n} \iff T$


Dado uma transformação matriz $A_{m \times n}$

$T(x) = A.x$

---

Agora dado uma transformação $T$ ($T:\mathbb{R}^n \to \mathbb{R}^m$)

Usamo as as bases dos 2 espaços (domínio $B_D$ e contradomínio $B_I$) para encontrar a matriz A associada a transformação T.
Verificamos como cada componente da base do domínio é transformada:

$T:\mathbb{R}^n \to \mathbb{R}^m$

$B_D = \{e_1, e_2, ..., e_n\} \implies X \in \mathbb{R}^n \implies X = x_1 e_1 + x_2 e_2 + ... + x_n e_n$ 

$B_I = \{e_1, e_2, ..., e_m\} \implies T(X) = x_1 T(e_1) + x_2 T(e_2) + ... + x_n T(e_n)$, onde $T(e_i) \in \mathbb{R}^m$

E $T(e_i)$ mostra como cada base de $\mathbb{R}^n$ é levada em $\mathbb{R}^m$

Logo, $T(e_i) = a_{1i} e_1 + a_{2i} e_2 + ... + a_{mi} e_m$ 

$A = \begin{bmatrix}
    a_{11} & a_{12} & a_{13} & \dots  & a_{1n} \\
    a_{21} & a_{22} & a_{23} & \dots  & a_{2n} \\
    \vdots & \vdots & \vdots & \ddots & \vdots \\
    a_{m1} & a_{m2} & a_{m3} & \dots  & a_{mn}
\end{bmatrix}_{m \times n}$

---
Algumas definições de matrizes

**Upper-triangules** $A = (a_{ij}) \implies a_{ij}=0$ se $i < j$

**Lower-triangules** $A = (a_{ij}) \implies a_{ij}=0$ se $i > j$

Tridiagonal $a_{ij}=0$ se $ |i - j| > 1$, aumentando 1 temos outras matrizes d-diagonais chamadas **matrizes em banda (*band matrices*)**

As matrizes triangulares têm no máximo $1/2 n^2 - 1/2 n$ ($n^2 - \sum_{i=1}^{n-1}i$) elementos não nulos.

Já as matrizes em banda têm no máximo $n + 2np - p^2 - p$ ($n + 2(n-1) + 2(n-1) + ... + 2(n-p) = n - 2\sum_{i=n-p}^{n-1}i$) elementos não nulos.


**Forma Hessenberg**

$a_{ij} = 0$ se $i > j +1$ (triangular superior com a primeira subdiagonal diferente de zero)

$H = \begin{bmatrix}
    a_{11} & a_{12} & a_{13} & \dots  & a_{1(n-1)}   & a_{1n} \\
    a_{21} & a_{22} & a_{23} & \dots  & a_{2(n-1)} & a_{2n} \\
    0 & a_{32} & a_{33} & \dots  & a_{3(n-1)} & a_{3n} \\
    0 & 0 & a_{43} & a_{44} & \dots  & a_{4n} \\
    \vdots & \vdots & \vdots & \vdots & \ddots & \vdots \\
    0 & 0 & 0 & \dots  & a_{(m-1)(n-1)} & a_{(m-1)n} \\
    0 & 0 & 0 & \dots  & a_{m(n-1)} & a_{mn} 
\end{bmatrix}_{m \times n}$


Um fato interessante sobre a forma Hessenberg é que é possível transformar qualquer matriz $A$ em sua forma Hessemberg num número finito de passos. Mas não dá para transformar qualquer matriz $A$ em uma triangular superior ou inferior. Pela teoria de Galois, não há fórmula fechada para encontrar as raízes de um polinômio de grau maior que 5. Logo, se fosse possível transformar qualquer matriz em uma triangular num número finito de passos, as raízes do polinômio característico da matriz $A$ seriam os elementos da diagonal da triangular, e seria possível encontrá-los com $n$ passos. Ou seja, esse seria um método direto (fórmula fechada) para encontrar as raízes de um dado polinômio.

**Traço e determinante**

Traço, $T_r(A) = \sum_{i=1}^{n} a_{ii}$

Determinante, $det(A) = \sum_{P \in \pi} Sign(P) . a_{1P_1} . a_{2P_2} . ... . a_{nP_n}  $, onde 

$\pi$  é o conjunto de todas as permutações de $\{1,2,3,...,n\}$,

$
Sign(P) = 
\left\{
    \begin{array}{l} 
        +1, \text{se o número de mudanças é par} \\ 
        -1, \text{se o número de mudanças é ímpar}
    \end{array}
\right.
$

Por exemplo, se $P = \{ 1,2,3, ..., n \}$, $Sign(P) = +1$ pois é o conjunto original e não houve mudança. Mas $P = \{ 1,3,2, ..., n \}$, $Sign(P) = -1$ pois houve uma mudança ($2$ e $3$).


Propriedades do determinante
1. $det(AB) = det(a) . det(B)$
2. $det(A+B) \neq det(a) + det(B)$
3. $det(\alpha A) = \alpha^n det(A)$
4. $det(A) = a_{11}.a_{22}.a_{33}...a_{nn}$, se $A$ é matriz triangular
5. $det(A^{-1}) = 1/det(A)$
6. $det(A^p) = [det(A)]^p, \forall p \in  \mathbb{Z}$
